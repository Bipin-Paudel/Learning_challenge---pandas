# Melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({ "name":["a","b","c","d" ,"a" , "b", "a", "b" , "a", "c","c"],
                      "s_1":[13,14,12,13,14,15,23,25,16,10,34], 
                      "s_2":[23,24,25,26,27,28,29,30,25,34,35] })
```

```python
pd.melt(var)
```

```python
pd.melt(var, var_name="python", value_name='one', id_vars='s_1' )
```

# pivot()

```python
var = pd.DataFrame({
    "days": [1, 2, 3, 4, 5, 6],
    "st_name": ["a", "b", "c", "a", "b", "c"],
    "eng": [10, 12, 14, 15, 16, 12],
    "maths": [17, 18, 19, 13, 14, 16]
})

var
```

```python
var.pivot(index='days', columns='st_name')
```

```python
var.pivot(index='days', columns='st_name', values='eng')
```
