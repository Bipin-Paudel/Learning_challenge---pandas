# Delete & Inserts function

```python
import pandas as pd

var = pd.DataFrame({'A':[2,3,4,5,6],'B':[1,5,7,8,9]})

```

```python
var.insert(0,'A_2', [2,7,4,5,9] )
print(var)
```

```python
var.insert(2,"b", var['B'])
var
```

```python
var['python12']= var['A'][:3]
var
```
